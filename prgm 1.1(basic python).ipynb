```python
#numeric datatypes
int=10
x=20.5
y=3 +4j

#datatypes and its values
print("integer :",int,type(int))
print("float :",x,type(x))
print("complex :",y,type(y))

#mathatematical operations
int +=5
x *=3
y -=1 +6j

#print updated values
print("updated value of int :",int)
print("updated value of Float :",x)
print("updated value of complex :",y)

#Boolean
t=True
f=False

#boolean operation
print("Boolean AND :",t and f)
print("Boolean OR :",t or f)
print("Boolean NOT :", not f)

#String
s="hello Morning"
d="  World     "

print(s+d)
#length
print("length of string",len(s))
#format
print("format String :",f"Goood {s} fine {d}!")

#string methods
print("capitalized :",s.capitalize())
print("Uppercase :",s.upper())
print("justified :",s.ljust(20,'*'))
print("centerd :",s.center(20,'-'))
print("Replaced :",s.replace("hello","Good"))
print("Striped :", d.strip())

#dictionary
person={
    "name":"Yadu",
    "age":23
    
}
print("Dictionary:",person )
print("Name :",person["name"])
#add and update dictionary
person["city"] = "TVM"
person["age"] = 22
print("updated Dictionary:",person )

#set
fruits ={"apple","orange","cherry"}
print("set :",fruits)

#add and remove from set
fruits.add("banana")
fruits.remove("apple")
print("updated set :",fruits)

#function
def add(a, b):
    return a + b
print("sum(using function)",add(5 ,5))
print("\n---------------------------\n")

# Find the largest in the list
# z = input("Enter a list of numbers separated by commas: ")
# x = list(map(int, z.split(',')))
z=[10,20,48,34,87,10]
max_value = max(z)
print("List of numbers:", z)
print("Largest number in the list:", max_value)

#remove duplicate value
x=list(set(z))
print("list of unique numbers",x)

#merge two dictionary
d1={'name': 'Yadu','age': 23}
d2={'city': 'clt','id':29}
md={**d1,**d2}
print("Dictionary 1 :",d1)
print("Dictionary 2 :",d2)
print("Merged Dictionary  :",md)

#Common Member
list1=[1,2,3,4,5,6]
list2=[5,6,7,8,9,10]
print("print list 1",list1)
print("print list 2",list2)
result=False
for x in list1:
    for y in list2:
        if x==y:
            result=True
            print(result)
if result:
    print("List have at least one common member")
else:
    print("List do not have any common member")
```
