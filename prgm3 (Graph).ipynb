```python
import matplotlib.pyplot as plt
import numpy as np

# Increase figure size
plt.figure(figsize=(16, 14))  # Increased height to provide more space

# Graph 1: Line Plot
x = np.array([1, 5])
y = np.array([2, 10])

plt.subplot(4, 3, 1)
plt.title('Line Plot')
plt.xlabel("X label", color="blue")
plt.ylabel("Y label", color="blue")
plt.plot(x, y, 'o', color="red")

# Graph 2: Multiple Line Plots
y1 = np.array([2, 4, 6, 8])
y2 = np.array([5, 3, 7, 6])

plt.subplot(4, 3, 2)
plt.plot(y1, color="green", marker='s', label='Series 1')
plt.plot(y2, linewidth=2, color="purple", linestyle="--", label='Series 2')
plt.title("Multiple Line Plots")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()

# Graph 3: Scatter Plot with Different Marker Styles
ypoints = np.array([1, 4, 2, 5])

plt.subplot(4, 3, 3)
plt.plot(ypoints, marker='^', ms=12, mec="blue", mfc="cyan")
plt.title("Scatter Plot")
plt.xlabel("Index")
plt.ylabel("Value")

# Graph 4: Line Plot with Different Data
x = np.array([50, 55, 60, 65, 70])
y = np.array([150, 160, 170, 180, 190])

plt.subplot(4, 3, 4)
plt.title("Line Plot")
plt.plot(x, y, color='magenta')
plt.grid()

# Graph 5: Scatter Plot with New Data
x = np.random.randint(100, size=(50))
y = np.random.randint(100, size=(50))
colors = np.random.rand(50)  # Color map based on random floats
sizes = 100 * np.random.rand(50)  # Increased sizes for larger dots

plt.subplot(4, 3, 5)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='plasma')
plt.title("Scatter Plot")

# Graph 6: Horizontal Bar Plot
x = np.array(["E", "F", "G", "H"])
y = np.array([12, 5, 8, 20])

plt.subplot(4, 3, 6)
plt.barh(x, y, color="teal")
plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Categories")

# Graph 7: Pie Chart with Different Data
y = np.array([40, 20, 25, 15])
mylabels = ["Oranges", "Pears", "Peaches", "Plums"]
myexplode = [0.1, 0, 0, 0.2]

plt.subplot(4, 3, 7)
plt.pie(y, labels=mylabels, explode=myexplode, startangle=140, shadow=True, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Pie Chart")
# Adjust the legend position to be inside and transparent
plt.legend(title="Fruit Types:", loc='upper left', bbox_to_anchor=(0.75, 0.75), framealpha=0.5)

# Graph 8: Histogram
data = np.random.randn(1000)  # Normally distributed data

plt.subplot(4, 3, 8)
plt.hist(data, bins=30, color="orange", edgecolor="black")
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Graph 9: Simple Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.subplot(4, 3, 9)
plt.bar(categories, values, color='skyblue')
plt.title("Simple Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")

# Overall title
plt.suptitle("Graphs", fontsize=20)

# Adjust subplot spacing
plt.subplots_adjust(left=0.1, right=0.85, top=0.92, bottom=0.05, wspace=0.3, hspace=0.4)

# Show plot
plt.show()
```
