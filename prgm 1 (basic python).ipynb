```python
#numeric datatypes
int=10
x=20.5
y=3 +4j

#datatypes and its values
print("integer :",int,type(int))
print("float :",x,type(x))
print("complex :",y,type(y))

#mathatematical operations
int +=5
x *=3
y -=1 +6j

#print updated values
print("updated value of int :",int)
print("updated value of Float :",x)
print("updated value of complex :",y)

#Boolean
t=True
f=False

#boolean operation
print("Boolean AND :",t and f)
print("Boolean OR :",t or f)
print("Boolean NOT :", not f)

#String
s="hello Morning"
d="  World     "

print(s+d)
#length
print("length of string",len(s))
#format
print("format String :",f"Goood {s} fine {d}!")

#string methods
print("capitalized :",s.capitalize())
print("Uppercase :",s.upper())
print("justified :",s.ljust(20,'*'))
print("centerd :",s.center(20,'-'))
print("Replaced :",s.replace("hello","Good"))
print("Striped :", d.strip())

#dictionary
person={
    "name":"Yadu",
    "age":23
    
}
print("Dictionary:",person )
print("Name :",person["name"])
#add and update dictionary
person["city"] = "TVM"
person["age"] = 22
print("updated Dictionary:",person )

#set
fruits ={"apple","orange","cherry"}
print("set :",fruits)

#add and remove from set
fruits.add("banana")
fruits.remove("apple")
print("updated set :",fruits)

#function
def add(a, b):
    return a + b
print("sum(using function)",add(5 ,5))
```
